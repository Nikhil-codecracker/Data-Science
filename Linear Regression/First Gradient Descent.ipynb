# Gradient Descent (in general)

##### Helps to find a local minima in any given function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

#### ***GOAL*** Given a Function f(x) ,we want to find the value of x that minimize f

### Visualisation 

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel('f(x)')
plt.xlabel('X')
plt.show()
```

```python
x = 0
lr = 0.1
error = []
plt.plot(X,Y)

# take 50 steps in the downward direction
for i in range(50):
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Plot the values of error

plt.plot(error)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
#Generate Datasets
X,y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

Y = y.reshape(-1,1)
print(X.shape)
print(Y.shape)
```

```python
#Normalisation
u = np.mean(X)
std = np.std(X)
X = (X-u)/std

#Visualise
plt.figure()
plt.scatter(X,Y)
plt.title("Visualise data")
plt.show()
```

```python
ones =  np.ones((X.shape[0],1))
X_ = np.hstack((X,ones))
print(X_.shape)
print(X_[:5,:])
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def getThetaClosedForm(X,Y):
    
    Y = np.mat(Y)
    first_part =np.dot(X.T,X)
    second_part = np.dot(X.T,Y)
    
    theta = np.linalg.pinv(first_part)*second_part
    return theta
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label='prediction')
plt.title("Visualise data")
plt.legend()
plt.show()
```

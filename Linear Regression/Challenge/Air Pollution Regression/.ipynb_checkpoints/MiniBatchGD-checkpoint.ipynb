```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
%matplotlib inline
```

```python
X,y = make_regression(n_samples=10000,n_features=20,n_informative=20,noise=10,random_state=1)
print(X.shape,y.shape)
```

```python
u = np.mean(X,axis=0)
std = np.std(X,axis=0)
X = (X-u)/std
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X.shape, y.shape)
```

```python
def hypothesis(X,theta):
    return np.dot(X,theta)

def error(X,y,theta):
    e = 0.0
    m = X.shape[0]
    y_ = hypothesis(X,theta)
    e = (np.sum((y-y_)**2))
    return e/m
def gradient(X,y,theta):
    y_ = hypothesis(X,theta)
    grad = np.dot(X.T,(y_-y))
    m = X.shape[0]
    return grad/m
def gradient_Descent(X,y,learning_rate = 0.01,max_iters = 300):
    n = X.shape[1]
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_iters):
        e = error(X,y,theta)
        error_list.append(e)
        
        grad = gradient(X,y,theta)
        theta = theta - learning_rate*grad
    return theta,error_list
```

```python
def batch_GD(X,y,batch_size = 20, learning_rate = 0.01,max_iter = 1):
    n = X.shape[1]
    theta = np.zeros((n,))
    error_list = []
    
    data = np.hstack((X,y.reshape(-1,1)))
    
    for i in range(max_iter):
        
        #Mini Batch
        np.random.shuffle(data)
        m = data.shape[0]
        total_batches = m//batch_size
        
        for i in range(total_batches):
            batch_data = data[i*batch_size : (i+1)*batch_size,:]
            x_batch = batch_data[:,:-1]
            y_batch = batch_data[:,-1]
            batch_grad = gradient(x_batch,y_batch,theta)
            theta = theta - learning_rate*batch_grad
            e = error(x_batch,y_batch,theta)
            error_list.append(e)
        
    return theta,error_list
```

```python
theta,error_list = batch_GD(X,y)

final_error = error_list[-1]
print(final_error)
plt.style.use('seaborn')
plt.figure()
plt.plot(np.arange(len(error_list)),error_list)
plt.show()
```

```python
theta,error_list = gradient_Descent(X,y)

final_error = error_list[-1]
print(final_error)
plt.figure()
plt.plot(np.arange(len(error_list)),error_list)
plt.show()
```

# Mini Batch Gradient Descent
- Faster Convergence
- Noisy Updates help avoid local Minima
- Use Vectorisation
- No need to load Entire dataset at the same time

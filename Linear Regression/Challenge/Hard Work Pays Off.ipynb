```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv("Linear_X_Train.csv")
y_train = pd.read_csv("Linear_Y_Train.csv")
x_test = pd.read_csv("Linear_X_Test.csv")
print(x_train.shape,y_train.shape,x_test.shape)
```

```python
x_train = x_train.values
y_train = y_train.values

u = x_train.mean()
std = x_train.std()
x_train = (x_train-u)/std
print(x_train.shape)
```

```python
def hypothesis(x,theta):
    y_pred = theta[0] + theta[1]*x
    return y_pred

def gradient(x,y,theta):
    m = x.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
        y_pred = hypothesis(x[i],theta)
        y_train = y[i]   
        grad[0] += (y_pred-y_train)
        grad[1] += (y_pred-y_train)*x[i]
    return grad/m

def error(x,y,theta):
    m = x.shape[0]
    total_error = 0.0
    for i in range(m):
        y_pred = hypothesis(x[i],theta)
        total_error += (y_pred-y[i])**2
    return total_error/m

def gradientdescent(x,y,learning_rate=0.1,max_itr = 1000):
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    m = x.shape[0]
    for i in range(max_itr):
        
        e = error(x,y,theta)
        error_list.append(e)
        grad = gradient(x,y,theta)
        
        theta[0] = theta[0] - learning_rate*grad[0]
        theta[1] = theta[1] - learning_rate*grad[1]
        theta_list.append(theta)
        
    return theta,error_list,theta_list
```

```python
theta,error_list,theta_list = gradientdescent(x_train,y_train)
print(theta)
```

```python
y_pred =hypothesis(x_test.values,theta)
print(y_pred)
```

```python
df = pd.DataFrame(y_pred,columns = ["y"])
df.to_csv('predict.csv',index = False)
```

```python
y_pred_train = hypothesis(x_train,theta)
print(y_pred_train.shape)
```

```python
def score(y_pred,y):
    num = np.sum((y-y_pred)**2)
    denom = np.sum((y-y.mean())**2)
    score = 1-num/denom
    return score
```

```python
score(y_pred_train,y_train)
```

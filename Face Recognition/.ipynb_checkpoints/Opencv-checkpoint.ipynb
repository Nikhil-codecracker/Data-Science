```python
import cv2
```

```python
img = cv2.imread('image.jpg')
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```

### Opencv reads image in BRG format by default rather than RGB

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #to convert BRG to RGB
plt.imshow(img2)
plt.show()
```

```python
print(img)
```

```python
cv2.imshow('Scenery',img) # to mention a heading, using this we can read
                          # iamge in the same format(RGB,BGR,GRB,etc.)
cv2.waitKey(0) #it tells how many millisec to before destroying the image
cv2.destroyAllWindows()#window, 0 is for infinite time
```

```python
gray = cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
cv2.imshow('Scenery',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

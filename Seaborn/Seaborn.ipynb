```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
sns.barplot(x='sex' ,y='total_bill' ,data=tips,estimator=np.std)
```

```python
sns.countplot(x = 'sex' , data =tips)
```

```python
sns.countplot(x = 'time', data = tips)
```

```python
sns.boxplot(x='day' ,y='total_bill' ,data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y = 'total_bill', data=tips)
```

```python
sns.violinplot(x='day',y = 'total_bill', data=tips, hue = 'sex', split = True)
```

```python
sns.distplot(tips['total_bill'], kde = False,bins = 30) 
#kde = False to remove the line representing shape of the graph
#bins for size of each bar for precision
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips)
# produces distplot for both tips and total_bill
#kind tells the type of marking for the relationship between x and y by default it is scatter
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips,kind = 'hex')
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips,kind ='kde' )
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips,kind = 'reg')
```

```python
sns.pairplot(tips,hue='sex')
#gives the replational graph between all the features
#hue makes different plots in same graph for the feature mentioned
```

```python
tips_corr = tips.corr()
#it taes all the non categorical variable and gives a relationship between them 
```

```python
sns.heatmap(tips_corr,annot=True)
#Used to plot graph for data which have variables(heading) for both x and y 
#Light colour means high correlation
#annot = True also shows the numbers
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot_table(index = 'month',columns='year',values = "passengers")
flights_pivot
#it is used to change the dataset such that we have index for row and columns for columns compared according to the passengers
```

```python
sns.heatmap(flights_pivot)
#brightest part means the most and lightest part means the least
```

```python
sns.heatmap(flights_pivot, cmap = 'coolwarm')
```

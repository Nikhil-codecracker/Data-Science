- Optimsation:  Gradient Descent/ Socastic

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples=500,n_features=2,centers=2,random_state=11)
print(X.shape,Y.shape)
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.show()
```

### Model and Helper Functions 

```python
def sigmoid(z):
    return (1.0)/(1+np.exp(-z))
```

```python
z = np.array([1,2,3,4,5])
sigmoid(z)
```

### Implement Perceptron Learning Algorithm
- Learn the weights
- Reduce the loss
- Make the predictions

```python
def predict(X,w):
    return sigmoid(np.dot(X,w))

def loss(X,Y,w):
    y_pred = predict(X,w)
    loss = np.mean(-(Y*np.log(y_pred) + (1-Y)*np.log(1-y_pred)))
    return loss

def update(X,Y,w,learning_rate):
    y_pred = predict(X,w)
    grad = np.dot(X.T,y_pred-Y)
    
    m = X.shape[0]
    
    w = w - learning_rate*grad/(float(m))
    return w
    
def train(X,Y,learning_rate=0.5,maxItr=100):
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    w = np.zeros(X.shape[1])
    
    for i in range(maxItr):
        w = update(X,Y,w,learning_rate)
        
        if i%10==0:
            l = loss(X,Y,w)
            print("Epoch %d Loss %.4f"%(i,l))
            
    return w
```

```python
weights = train(X,Y,maxItr=1000)
```

### Prepration Implementation - Part-II 
- Make Prediction
- Visualise Decision Surface
- Linear vs Non-Linear classification

```python
def getprediction(X_test,weights,labels=True):
    if X_test.shape[1]!=weights[0]:
        ones = np.ones((X_test.shape[0],1))
        X_test = np.hstack((ones,X_test))
    
    probs = predict(X_test,weights)
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs>=0.5]=1
        return labels
```

```python
x1 = np.linspace(-8,2,10)
print(x1)

x2 = -(weights[0]+weights[1]*x1)/weights[2]
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.plot(x1,x2,c='red')
plt.show()
```

```python
# Find the predictions
Y_ = getprediction(X,weights,labels=True)
print(np.sum(Y_==Y)/Y.shape[0])
```

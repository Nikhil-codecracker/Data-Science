```python
from urllib.request import urlopen
```

```python
jokes_data = urlopen("http://www.icndb.com/the-jokes-2/")
```

```python
jokes_html = jokes_data.read()
print(jokes_html)
```

```python
jokes_data.close()
```

```python
from bs4 import BeautifulSoup as soup
```

```python
jokes_soup = soup(jokes_html,'html.parser')
print(jokes_soup)
```
